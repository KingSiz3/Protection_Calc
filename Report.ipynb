## Introduction 
A MV/HV cable subjected to a L-G fault will have the short circuit current flow in the earthed cable wire screen until fault interruption. To calculate the permissible thermal short circuit current for cable conductors and wire screens the methods and formulas from IEC60949/1988 are used. In short terms, heating of the current carrying conductor is considered adiabatic and hence yielding a first order differential equation \eqref{eq1}. The general solution is given \eqref{eq2}. Note that the general solution considers only one current carrying component. 

$$
\begin{equation}
    I_{AD}^2 \cdot R' = S \cdot 10^{-6} \cdot \frac{d\theta(t)}{dt}, \quad\quad
    R' = \frac{\rho_{20} \frac{\theta(t)+\beta}{20+\beta} }{S \cdot 10^{-6}}
    \label{eq1} \tag{1}
\end{equation}
$$
<br>
$$
\begin{equation}
    I_{AD}^2 t = \cdot K^2 S^2 \ln \left( \frac{\theta_f + \beta}{\theta_i + \beta} \right), \quad\quad
    K = \sqrt{\frac{\sigma_c (\beta+20) \cdot 10^{-12}}{\rho_{20}}}
    \label{eq2} \tag{2}
\end{equation}
$$
Where,
<br>
<center>
$\small\begin{array} {l l l}
I_{AD} & [\text{A}] & : & \text{ the adiabatic short circuit current} \\
\theta(t)  & [^{\circ}\text{C}] & : & \text{the temperature of the current conducting material as a function of time} \\
\theta_i  & [^{\circ}\text{C}] & : & \text{the initial temperature, } t = 0 \\
\theta_f  & [^{\circ}\text{C}] & : & \text{the final temperature, } \\
R'  & [\Omega / \text{m}] & : & \text{the resistance of the component as a function of temperature} \\
\rho_{20}  & [\Omega m] & : & \text{the electrical resistance of the current conducting material at 20 }^{\circ}\text{C} \\
\beta & [\text{K}] & : & \text{reciprocal of temperature coefficient of resistance} \\ 
      &            &   & \text{of the current carrying component at 0 }^{\circ}\text{C} \\ 
\sigma_c & [\text{J/K.m}^3] & : & \text{the volumetric specific heat of the current carrying component at 20 }^{\circ}\text{C} \\
S & [\text{mm}] & : & \text{the cross-section area of component} \\
t & [\text{s}] & : & \text{time} \\
K & [\text{As}^{1/2}/\text{mm}^2] & : & \text{a constant depending on the current carrying component} \\
\end{array}$
</center>

For wire screens and sheaths IEC60949/1988 introduces a non-adiabatic factor $\varepsilon$, which appreciate the fact that some heat will be transfer to the surrounding materials. 

In year 2008 an amendment 1 was added to IEC60949, which states below assumption:

*Where there are a number of components within the cable that are connected in parallel, e.g. screen/sheath/armour, such that they will share the fault current, it is sufficiently accurate to consider that the ratio between the currents in any two components is equal to the inverse ratio of their respective resistance. Each component is expected to achieve a different temperature. ...*
In other words, when calculating the permissible short circuit current for cable with both a copper wire screen and a radial water barrier (aluminum sheath) one could consider both as current carrying components, see model of an arbitrary submarine cable in figure 1. 

Sadly, the amendment does not provide any solution for this consideration. For each considered current conducting component we get an additional coupled differential equation. An appropriate solution/estimation to the system of differential equations can be found in several ways. The paper [[3](#ref3)] introduces a general analytical expression to the problem, solving for n-components. In section [Appendix B: Calculations](#AppendixB), I have applied these analytical expression and applied an iterative method for validation. In addition, I have applied the regular expression in IEC60949/1988. All methods are applied on the cable in [Appendix A: Cable data](#AppendixA).

## Appendix B: Calculations

### Base case - IEC60949/1988 
In this section the non-adiabatic short circuit current is calculated for the cable data using the expressions in IEC60949/1988.

```python
I_AD, I_NAD, epsilon, M = calc_screen_thermal_scc_non_adiabatic( \
                                                            con_mat = cu,
                                                            ins_mat_1 = xlpe,
                                                            ins_mat_2 = xlpe,
                                                            S = S,
                                                            theta_f = theta_f,
                                                            theta_i = theta_i,
                                                            t = t,
                                                            delta = delta,
                                                            F = F)
```

Using the general solution from IEC60949/1988, the energy needed to raise the copper-wire screen temperature from $\theta_{sc, i}$ to $\theta_{sc, f}$ is calculated:

```python
md("$ \
    \\begin{equation} \
       I_{AD}^2 t = K^2_{cu} S^2_{sc} \ln{ \left( \\frac{\\theta_{sc,f} + \\beta_{cu}}{\\theta_{sc, i} + \\beta_{cu}}\\right)} = %s \\text{  [J]} \
       \label{eq3} \\tag{3} \
    \\end{equation} \
   $"%format(I_AD**2*t, '.2f'))
```

At a given short circuit duration, the equvivalent permissible adiabatic short circuit current $I_{AD}$ can be calculated:

```python
md("$ \
    \\begin{equation} \
       I_{AD} = \sqrt{K^2_{cu} S^2_{sc} \ln{ \left( \\frac{\\theta_{sc,f} + \\beta_{cu}}{\\theta_{sc,i} + \\beta_{cu}}\\right)} \cdot t^{-1}} = %s \\text{  [A]} \
       \label{eq4} \\tag{4} \
    \\end{equation} \
   $"%format(I_AD, '.2f'))
```

Next the non-adiabatic factor $\varepsilon$ is calculated. <br>
$\sigma_2/\sigma_3$ and $\rho_2/\rho_3$ are the volumetric specifc heat and thermal resistivities repectively of the adjacent non-metallic materials.

```python
md("$ \
    \\begin{equation} \
       M = \\frac{\\left(\sqrt{\\frac{\\sigma_2}{\\rho_2}} + \sqrt{\\frac{\\sigma_3}{\\rho_3}} \\right)}{2 \\sigma_1 \\delta \cdot 10^{-3}} F = %s \\text{  [s}^{-1/2}] \
       \label{eq5} \\tag{5} \
    \\end{equation} \
   $"%format(M, '.4f'))
```

```python
md("$ \
    \\begin{array}{ll} \
       \\varepsilon & = 1 + 0.61 \cdot M \cdot t^{{1 / 2}} - 0.069 \cdot \left(M \cdot t^{{1 / 2}} \\right) ^2 + 0.0043 \cdot \left(M \cdot t^{1 / 2} \\right)^3 \\\ \
       & = %s \\text{  [s}^{-1/2}] \
       \label{eq6} \\tag{6} \
    \\end{array} \
   $"%format(epsilon, '.4f'))
```

At last the non-adiabatic short circuit current is given:

```python
md("$ \
    \\begin{equation} \
       I_{NAD} = I_{AD} \cdot \\varepsilon = %s \\text{  [A]} \
       \label{eq7} \\tag{7} \
    \\end{equation} \
   $"%format(I_NAD, '.2f'))
```

### Case 1 -  IEC60949 AMD1 Analytical expression 
In the section the permissible adiabatic short circuit is calculated for the cable using the analytical solution for the n-component problem proposed in the white paper [[3](#ref3)].

#### Permissible thermal short circuit (adiabatic)

```python
from func_IEC60949 import component, sys, iterate, bisection_search
screen = component( name = 'screen',
                    mat = cu,
                    S = 35,
                    theta_i = theta_i,
                    theta_max = theta_f,
                    d = 48.4,
                    P = 387)

sheath = component( name = 'sheath',
                    mat = al,
                    S = 33,
                    theta_i = theta_i,
                    theta_max = theta_f)
# Analytical
s = sys([screen, sheath])
s.calc_S(screen, I_NAD**2*t)
# Iterative
ntheta, nE = iterate(components = [screen, sheath], Eref = s.E, dt = 0.001)
Sn = bisection_search([screen, sheath], screen, I_NAD**2*t)
```

From the inital temperature the initial resistances can be determined:

```python
md("$ \
    \\begin{equation} \
       R^{'}_{sc,i} = \\frac{\\rho_{cu,20} \\frac{\\theta_{sc,i}+\\beta_{cu}}{20+\\beta_{cu}} }{S_{sc} \\cdot 10^{-6}} = %s \\text{  }[\\Omega / \\text{m}] \
       \label{eq8} \\tag{8} \
    \\end{equation} \
   $"%format(screen.R_i, '.5g'))
```

```python
md("$ \
    \\begin{equation} \
       R^{'}_{sh,i} = \\frac{\\rho_{al,20} \\frac{\\theta_{sh,i}+\\beta_{al}}{20+\\beta_{al}} }{S_{sh} \\cdot 10^{-6}} = %s \\text{  }[\\Omega / \\text{m}] \
       \label{eq9} \\tag{9} \
    \\end{equation} \
   $"%format(sheath.R_i, '.5g'))
```

Next we need to determine which component is going to reach their maximum allow temperature $\theta_{max}$ first.

Screen:

```python
md("$ \
    \\begin{equation} \
       \\left[ \\frac{K_{sc} G_{sc}\\rho_{cu,20}\cdot10^6}{\\left( 20 + \\beta_{cu} \\right)} \\right]^2 \\left[ \\left(\\theta_{sc,max} + \\beta_{cu}  \\right)^2 - \\left(\\theta_{sc,i} + \\beta_{cu}  \\right)^2 \\right]= %s \\text{  }[\\text{J}\cdot\\Omega/\\text{m}^4] \
       \label{eq10} \\tag{10} \
    \\end{equation} \
   $"%format(screen.limit, '.2f'))
```

Sheath:

```python
md("$ \
    \\begin{equation} \
       \\left[ \\frac{K_{sh} G_{sh}\\rho_{20}\cdot10^6}{\\left( 20 + \\beta_{al} \\right)} \\right]^2 \\left[ \\left(\\theta_{sh,max} + \\beta_{al}  \\right)^2 - \\left(\\theta_{sh,i} + \\beta_{al}  \\right)^2 \\right]= %s \\text{  }[\\text{J}\cdot\\Omega/\\text{m}^4] \
       \label{eq11} \\tag{11} \
    \\end{equation} \
   $"%format(sheath.limit, '.2f'))
```

Limiting component in this case is the <ins>screen</ins>. Hence, it will first reach its maximum permissible temperature of 250 $^{\circ}$C, and the final resistance $R^{'}_{sh, f}$ can then be calculated:

```python
md("$ \
    \\begin{equation} \
       R^{'}_{sc,f} = \\frac{\\rho_{20}\\frac{\\theta_{sc,max}+\\beta}{\\theta_{i}+\\beta}}{S\cdot10^{-6}} G = %s \\text{  }[\\Omega / \\text{m}] \
       \label{eq12} \\tag{12} \
    \\end{equation} \
   $"%format(s.components[0].R_f, '.6g'))
```

The final resistance of the other current carrying component (screen) can be calculated with below expression:

```python
md("$ \
    \\begin{equation} \
       R^{'}_{sc,f} = \\sqrt{\\frac{S_{sh}^2 K_{sh}^2}{S_{sc}^2 K_{sc}^2} \\left( R^{'2}_{sh,f} - R^{'2}_{sh,i} \\right) + R^{'2}_{sc,i}} = %s \\text{  }[\\Omega / \\text{m}] \
       \label{eq13} \\tag{13} \
    \\end{equation} \
   $"%format(s.components[1].R_f, '.6g'))
```

Finally, can the total permissible energy dissipation be calculated where $1$ is the sheath and $2$ is the screen: 

```python
md("$ \
    \\begin{equation}\
            E = I_{AD}^2 t = \sum_{j,k=1}^n S_j K_j S_k K_k \ln \\frac{R^{'}_{jf} S_{j} K_{j} + R^{'}_{kf} S_{k} K_{k}}{R^{'}_{ji} S_{j} K_{j} + R^{'}_{ki} S_{k} K_{k}}  = %s \\text{  [J}] \\\ \
       \label{eq14} \\tag{14} \
    \\end{equation} \
   $"%format(s.E, '.2f'))
```

The equivialent permissible short circuit current with repect to $t = 0.52$ is:

```python
md("$ \
    \\begin{equation}\
            I_{AD} = \\sqrt{E \\cdot t^{-1}}  = \\underline{\\underline{%s}} \\text{  [A}] \\\ \
       \label{eq15} \\tag{15} \
    \\end{equation} \
   $"%format((s.E/t)**(1/2), '.2f'))
```

#### Reduction in copper wire screen cross-section area
By considering both the copper wire screen and the aluminum sheath as conductors during the short circuit, the permissible thermal short cirucit energy has increased. Alternatively, we can decrease the copper wire cross-section area to match the permissible thermal short circuit energy calculated in the base case.

The paper [[3](#ref3)] offers a (long) analytical expression to calculate the cross-section area for the copper wire screen that matches a particular energy level.

```python
md("$ \
    \\begin{equation}\
            S_{sc} = \\frac{\\sqrt{\\Phi} - S_{sh} K_{sh} \\ln \\frac{R^{'}_{sc,uf} K_{sc} + R^{'}_{sh,uf} K_{sh}}{R^{'}_{sc,ui} K_{sc} + R^{'}_{sh,ui} K_{sh}}} {K_{sc} \\ln \\frac{R^{'}_{sc,uf}}{R^{'}_{sc,ui}}} = \\underline{\\underline{%s}} \\text{  }[\\text{mm}] \\\ \
       \label{eq16} \\tag{16} \
    \\end{equation} \
   $"%format(s.Sn, '.2f'))
```

```python
md("$ \
    \\begin{equation}\
           \\Phi = \\left( S_{sh} K_{sh} \\ln \\frac{R^{'}_{sc,uf} K_{sc} + R^{'}_{sh,uf} K_{sh}}{R^{'}_{sc,ui} K_{sc} + R^{'}_{sh,ui} K_{sh}} \\right)^2 + I_{AD}^2 t \\ln \\frac{R^{'}_{sc,uf}}{R^{'}_{sc,ui}}  \
           -\\ln \\frac{R^{'}_{sc,uf}}{R^{'}_{sc,ui}} S_{sh} K_{sh} S_{sh} K_{sh} \\ln \\frac{R^{'}_{sh,uf} K_{sh} + R^{'}_{sh,uf} K_{sh}}{R^{'}_{sh,ui} K_{sh} + R^{'}_{sh,ui} K_{sh}}\\\ \
       \label{eq17} \\tag{17} \
    \\end{equation} \
   $")
```

where $R^{'}_{ux}$ is the component DC resistance per unit length of cable and $x=\{i, f\}$ is the associated initial or final temperature.

```python
md("$ \
    \\begin{equation}\
            R^{'}_{ux} = R^{'}_{x} \cdot S = \\frac{\\rho_{20} \\frac{\\theta_{x}+\\beta}{20+\\beta} }{ 10^{-6}}\\\ \
       \label{eq18} \\tag{18} \
    \\end{equation} \
   $")
```

### Case 2 - Iterative method

To validate the results from the proposed analytical expression in [[3](#ref3)], I have implemented an iterative method.
The general solution in \eqref{eq2} can be rearrange to yield the final temperature $\theta_f$ as a function of time $t$.

```python
md("$ \
    \\begin{equation} \
       \\theta_f = \\left(\\theta_i + \\beta \\right) \\exp \\left[ \\frac{I_{AD}^2 t}{K^2 S^2} \\right] - \\beta \
       \\label{eq19} \\tag{19} \
    \\end{equation} \
   $")
```

To solve for a 2-component case we express the temperature $\theta$ for each component as a function of time $t$.

```python
md("$ \
    \\begin{align} \
       \\theta_{sc}(t) = \\left(\\theta_{i,sc} + \\beta_{sc} \\right) \\exp \\left[ \\frac{I_{sc}^2 t}{K_{sc}^2 S_{sc}^2} \\right] - \\beta_{sc} \quad\quad \
       \\theta_{sh}(t) = \\left(\\theta_{i,sh} + \\beta_{sh} \\right) \\exp \\left[ \\frac{I_{sh}^2 t}{K_{sh}^2 S_{sh}^2} \\right] - \\beta_{sh} \
       \\label{eq20} \\tag{20} \
    \\end{align} \
   $")
```

Next we calculate the current $I_{sc}, I_{sh}$ in each component as a function of the temperature $\theta$ 

```python
md("$ \
    \\begin{align} \
       I_{sc} = I_{AD} \cdot \\frac{R'_{sh}(\\theta_{sh})}{R'_{sh}(\\theta_{sh}) + R'_{sc}(\\theta_{sh})} \quad\quad \
       I_{sh} = I_{AD} \cdot \\frac{R'_{sc}(\\theta_{sh})}{R'_{sh}(\\theta_{sh}) + R'_{sc}(\\theta_{sh})} \
       \\label{eq21} \\tag{21} \
    \\end{align} \
   $")
```

Next step is to iterate over the expression in \eqref{eq20} and \eqref{eq21} with a predefine time step $dt$. Below is the pseudo code for a simple implementation. 
<br><br>


<center>
$\small\begin{array}{l}
    \text{Name } = \color{blue} {\text{iterate_E}} \\
    \begin{array} {l l}
        n  & : & \text{Number of current carrying components} \\
        E_{ref}  & : & \text{Reference energy} \\
        E & : & \text{Calculated energy} \\
        t & : & \text{time} \\
        dt  & : & \text{time step} \\
        R_j & : & \text{Electrical resistance} \\
        I_j  & : & \text{Component specific current} \\
        \theta_j & : & \text{Temperature} \\
        \theta_i  & : & \text{Initial temperature} \\
    \end{array} \\
    \begin{array} {l}
         \\
         \theta_1, \theta_2 \text{ . . . } \theta_n = \theta_i   \\
         \color{blue} {\text{for}} \text{  } t = \left[ 0, dt, 2dt \text{ . . . } t_{max} \right] : \\
         \quad \quad \color{gray} {\text{# Electrical resistance for each component with respect to the current temperature}} \\
         \quad \quad R_1, R_2  \text{ . . . } R_n = \color{blue} {\text{calc_Rj}}(\theta_n)  \\
         \quad \quad \color{gray} {\text{# Current distribution in each condcuting component}} \\
         \quad \quad I_1, I_2  \text{ . . . } I_n = \sqrt{E_{ref}} \cdot \frac{R_j^{-1}}{R_1^{-1}+R_2^{-1} + \text{ . . . } R_n^{-1}}  \\
         \quad \quad \color{gray} {\text{# Break if max temperature for a component is reached}} \\
         \quad \quad \color{blue} {\text{if}} \text{  } \left( \theta_1 \color{blue} {\text{ or }} \theta_2 \text{ . . . }  \color{blue} {\text{ or }} \theta_n  \right) >= \left( \theta_{1,max}, \theta_{2,max}, \text{ . . . } \theta_{n,max} \right)  : \\
         \quad \quad \quad \quad  \color{blue} {\text{break}}  \\
         \quad \quad \color{gray} {\text{# Break if the energy exceeds the refernces energy }E_{ref}} \\
         \quad \quad \color{blue} {\text{if}} \text{  } \left( I_1 + I_2 + \text{ . . . } +  I_n \right)^2 \cdot t >= E_{ref}  : \\
         \quad \quad \quad \quad  \color{blue} {\text{break}}  \\
         \quad \quad \color{gray} {\text{# Increment components temperature using the current temperature as initial,}} \\
         \quad \quad \color{gray} {\text{# the present current distribution and one time step } dt} \\
         \quad \quad \theta_1, \theta_2 \text{ . . . } \theta_n \mathrel{+}= \color{blue} {\text{calc_}\Delta\theta_j}(\left[ \theta_1, \theta_2 \text{ . . . } \theta_n \right], (\left[ I_1, I_2 \text{ . . . } I_n \right], t )   \\
         \color{blue} {\text{return}} \text{  } \theta_1 , \theta_2 , \left( E = (I_1+I_2)^2\cdot (t-dt) \right) \\

    \\

\end{array}$

#### Reduction in copper wire screen cross-section area
Using the result from method 1 ref \eqref{eq7} as reference energy, we can calculate the equivalent reduced copper wire cross-section area. This can be done in various ways. Here, I choose to use a simple bisection search algorithm to reduce the copper wire screen cross-section area until it matches the specific short circuit energy from the base case $E = I_{NAD}^2 \cdot t = (7974.01 \text{A})^2 \cdot 0.52 \text{s} = 33064114.45 [\text{J}]$. Below is the pseudo code for the implementation.

<center>
$\small\begin{array}{l}
    \text{Name } = \color{blue} {\text{bisection_search}} \\
    \begin{array} {l l} 
        E_{ref}  & : & \text{Reference energy} \\
        E & : & \text{Calculated energy} \\
        S_{sc, new} & : & \text{Screen cross-section area [mm]} \\
        a, b  & : & \text{end-point values for the cross-section area search} \\
        c & : & \text{mid-point} \\
        tol  & : & \text{tolerance for the solution} \\
    \end{array} \\
    \begin{array} {l}
         \color{blue} {\text{while}} \text{  } \left( |E-Eref| \color{blue} {\text{ or }} \frac{b-a}{2} \right) < tol  : \\
